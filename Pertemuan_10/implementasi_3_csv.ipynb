```python
#Import library yang dibutuhkan
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
#Menyiapkan data dan memanggil dataset
dataset = pd.read_csv('konsumen.csv')
dataset.keys()
```

```python
#Menampilkan 5 baris data pertama dari dataset
dataku = pd.DataFrame(dataset)
dataku.head()
```

```python
#Konversi Dataset ke Data Array
X = np.asarray(dataset)
print(X)
```

```python
#Menampilkan data Array ke dalam Scatter Plot
plt.scatter(X[:,0], X[:,1], label='True Position')
plt.xlabel("Gaji")
plt.ylabel("Pengeluaran")
plt.title("Grafik Penyebaran Data Konsumen")
plt.show()
```

```python
#Mengaktifkan K-Means dengan jumlahh K=2
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
```

```python
#Menampilkan nilai Centroid
print(kmeans.cluster_centers_)
```

```python
#Plot data point dan memvisualisasikan hasil klastering
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.xlabel("Gaji")
plt.ylabel("Pengeluaran")
plt.title("Grafik Hasil Klasterisasi Data Gaji dan Penegeluaran Konsumen")
plt.show()
```

```python
#Plot data point dan memvisualisasikan hasil klastering dengan centorid dari masing masing klaster
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], color='black')
plt.xlabel("Gaji")
plt.ylabel("Pengeluaran")
plt.title("Grafik Hasil Klasterisasi Data Gaji dan Penegeluaran Konsumen")
plt.show()
```
